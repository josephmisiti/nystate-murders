```python
import numpy as np
from ggplot import *
import os,datetime,random
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
from pandas.tools.plotting import scatter_matrix
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("data/data.csv")
COLOR='#A60628'
```

```python
data.head(2)
```

```python
COUNTIES = data.County.unique()
```

```python
grouped = data.groupby(["County","Year"]).Murder
murders = grouped.count()
murders.ix['Albany'].plot(kind='line',title)
```

```python
#murders.unstack(level=0).plot(kind='bar', subplots=True);
figsize(11, 20)
for i,c in enumerate(COUNTIES[0:]):
    sx = plt.subplot(len(COUNTIES)/4+1, 4, i)
    murders.xs(COUNTIES[i]).plot(kind='bar',color='#A60628')
    plt.title(COUNTIES[i])
    plt.ylim(0, murders.max())
plt.suptitle("Murders In Counties In NYS By Year", y=1.02, fontsize=14)
plt.tight_layout()
```

Although plotting it like this isnt particularly useful, you can begin to see which counties have a shitload of murders and which ones do not - of course this does not take into account the population density of the counties
